# Plot map of the provinces of the Netherlands with GeoPandas

To check if the coordinates of the geodata retrieved from the CBS (Dutch Central Bureau of Statistics) are correct I use geopandas to create the map. To render the correct view an crs-code is needed to display the coordinates correctly and with the *wkt* function from shapely the geodata can be loaded.

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt

pv = pd.read_csv("../data/cbs_provincie_2020_gegeneraliseerd.csv")
pv["geometry"] = pv["geometry"].apply(wkt.loads)
pv_geo = gpd.GeoDataFrame(pv, crs="EPSG:28992")
print(pv_geo.crs)
plot = pv_geo.plot()
plot.set_title("Oblique Stereographic")
```
